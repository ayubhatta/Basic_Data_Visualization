```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 	Write a python program to load data into pandas DataFrame 

```python
df = pd.read_csv('DataScienceSalaries_8b290669-a5e9-45bf-be72-d27add2eacae_93472_.csv')
df.head()
```

# 2. Write a python program to remove unnecessary columns i.e., salary and salary currency.

```python
df = df.drop(['salary', 'salary_currency'], axis=1)
df
```

# 3. Write a python program to remove the NaN missing values from updated dataframe

```python
df = df.dropna()
df
```

# 4. Write a python program to check duplicates value in the dataframe. 

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows    # prints all the duplicated values

# prints the given strings insted of all the values, if the rows are duplicate or not.
if duplicate_rows.empty:
    print("No duplicate rows found.")
else:
    print("Duplicate rows found.")
```

# 5. Write a python program to see the unique values from all the columns in the dataframe.		

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in column '{column}':")
    print(unique_values)
```

# 6. Rename the experience level columns.

```python
experience_level_mapping = {
    'SE': 'Senior Level/Expert',
    'MI': 'Medium Level/Intermediate',
    'EN': 'Entry Level',
    'EX': 'Executive Level'
}
```

```python
df['experience_level'] = df['experience_level'].replace(experience_level_mapping)
df
```

# 7. Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.

# 8. Write a Python program to calculate and show correlation of all variables. 

```python
# excluing non-numeric columns using function .select_dtypes
numeric_df = df.select_dtypes(include=['number'])

# Calculating correalations of all variables
correlation_matrix = numeric_df.corr()
correlation_matrix
```

# 9. Write a python program to find out top 15 jobs. Make a bar graph of sales as well.         

```python
job_counts = df['job_title'].value_counts().head(15)
job_counts
```

```python
plt.figure(figsize=(10, 6))
job_counts.plot(kind='bar', color='green')
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

# 10. Which job has the highest salaries? Illustrate with bar graph.	

```python
# Calculate the average salary for each job title
average_salary = df.groupby('job_title')['salary_in_usd'].mean()
average_salary
```

```python
plt.figure(figsize=(15, 6))
average_salary.plot(kind='bar', color='green')
plt.title('Top Jobs with Highest Salaries')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

# 11. Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
# Calculate the average salary for each experience level
average_salary_by_level = df.groupby('experience_level')['salary_in_usd'].mean()
average_salary_by_level
```

```python
plt.figure(figsize=(15, 8))  # Adjusted figure size
average_salary_by_level.plot(kind='bar', color='green')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90, ha='right')  # Rotating x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

# 12. Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.

```python
chosen_variable = 'salary_in_usd'
chosen_variable
```

```python
# Create a figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Histogram
axes[0].hist(df[chosen_variable], bins=10, color='green', edgecolor='black')
axes[0].set_title(f'Histogram of {chosen_variable}')
axes[0].set_xlabel(chosen_variable)
axes[0].set_ylabel('Frequency')

# Box plot
axes[1].boxplot(df[chosen_variable], vert=False, patch_artist=True)
axes[1].set_title(f'Box plot of {chosen_variable}')
axes[1].set_xlabel(chosen_variable)

# Adjust layout
plt.tight_layout()
plt.show()
```
